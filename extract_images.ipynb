```python
datasets = {'test1': {'gt': 'data/test/test1/gt',
                      'distorted': 'data/test/test1/input',
                      'trans_weather': 'results_orig/test1/Transweather/',
                      'qt': 'results_qt/test1/Transweather/',
                      'my': 'results/test1/Transweather/',
                      'images': ['im_0342_s90_a06.png', 'im_0324_s100_a05.png', 'im_0302_s85_a05.png', 'im_0345_s80_a06.png', 'im_0329_s80_a04.png', 'im_0307_s90_a06.png', 'im_0320_s100_a05.png']},
            'snowtest100k': {'gt': 'data/test/Snow100K-L/gt',
                             'distorted': 'data/test/Snow100K-L/input',
                             'trans_weather': 'results_orig/snowtest100k/Transweather/',
                             'qt': 'results_qt/snowtest100k/Transweather/',
                             'my': 'results/snowtest100k/Transweather/',
                             'images': ['beautiful_smile_02787.jpg', 'winter__street_00423.jpg', 'castle_00516.jpg','city_read_01302.jpg','snow_animal_00371.jpg', 'school_night_06876.jpg', 'city_read_04354.jpg',  'winter__street_07690.jpg', 'parking_lot_12047.jpg']},
            'raindroptest': {'gt': 'data/test/rain_drop_test/gt',
                             'distorted': 'data/test/rain_drop_test/input',
                             'trans_weather': 'results_orig/raindroptest/Transweather/',
                             'qt': 'results_qt/raindroptest/Transweather/',
                             'my': 'results/raindroptest/Transweather/',
                             'images': ['19_rain.png', '35_rain.png', '39_rain.png', '3_rain.png', '22_rain.png', '23_rain.png', '57_rain.png', '15_rain.png']}}
```

```python
for dataset in datasets.values():
    for imname in dataset['images']:
        base = os.path.basename(imname)[:-4]

        gt = cv2.imread(os.path.join(dataset['gt'], imname.replace('rain', 'clean')))
        cv2.imwrite(f"orig/{base}_gt.png", gt)

        distorted = cv2.imread(os.path.join(dataset['distorted'], imname))
        cv2.imwrite(f"orig/{base}_distorted.png", distorted)

        trans_weather = cv2.imread(os.path.join(dataset['trans_weather'], imname))
        cv2.imwrite(f"orig/{base}_trans_weather.png", trans_weather)

        if os.path.exists(os.path.join(dataset['qt'], imname)):
          qt = cv2.imread(os.path.join(dataset['qt'], imname))
          cv2.imwrite(f"orig/{base}_qt.png", qt)
        if os.path.exists(os.path.join(dataset['my'], imname)):
          my = cv2.imread(os.path.join(dataset['my'], imname))
          cv2.imwrite(f"orig/{base}_my.png", my)
```
